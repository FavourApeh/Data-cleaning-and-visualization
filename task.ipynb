```python
import pandas as pd
import numpy as np

#Read first dataset
data1 = pd.read_csv('dataset1.csv',encoding='ISO-8859-1')
data1.head()
```

```python
#Read second dataset
data2 = pd.read_csv('dataset2.csv')
data2.head()
```

```python
#Change column names for dataset1
column = ['Tiimestamp', 'Email', 'Interest_specialization', 'level_of_proficiency', 'tools', 'projects', 'portfolio', 'mentor_skills', 'improvement']
data1.columns = column
data1.head()
```

```python
#Change column names for dataset2
column = ['Timestamp', 'Email', 'Name', 'Phone_number', 'years_of_Data_experience', 'skillset', 'Group_No']
data2.columns = column
data2.head()
```

```python
#Join dataset1 and dataset2
data3 = pd.merge(data1, data2, on = 'Email', how='left')
data3
```

```python
#Drop some columns form the new dataset
data3.drop(data3.columns[[9,10,11,13,14]], axis=1, inplace=True)
data3
```

```python
#Rename the columns of the new dataset
columns = ['Timestamp', 'Email', 'Interest_specialization', 'level_of_proficiency', 'tools', 'projects', 'portfolio', 'mentor_skills', 'improvement', 'experience']
data3.columns = columns
data3
```

```python
#Convert the timestamp column to the data time object
data3['Timestamp'] = pd.to_datetime(data3['Timestamp'])
print(data3.info())
```

```python
#Resetting index
data3.reset_index(inplace=True)
data3.drop('index',axis=1,inplace=True)
```

```python
#Extracting month and day from timestamp
data3['month'] = data3['Timestamp'].dt.strftime('%b')
data3['day'] = data3['Timestamp'].dt.strftime('%d')
data3
```

```python
#Creating new dataframe that contains only data associated with data analysis or data science
new_data = data3.query("Interest_specialization == 'Data Analysis' or Interest_specialization == 'Data Science'")
new_data
```

```python
new_data.to_csv("cleaned_data.csv")
```
